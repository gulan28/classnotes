```python
import numpy as np
```

```python
training_data = [
    (1, 1),
    (2, 3),
    (4, 3),
]

def phi(x):
    return np.array([1, x])

def init_weights():
    return np.zeros(2)

def loss(w):
    return 1.0/len(training_data) * sum((w.dot(phi(x))- y)**2 for x, y in training_data)

def gradient_loss(w):
    return 1.0/len(training_data) * sum(2.0 * (w.dot(phi(x)) - y) * phi(x) for x, y in training_data)

def gradient_descent(Floss, Fgradient_loss, initial_weight_vector):
    epochs = 500
    learning_rate = 0.1
    w = initial_weight_vector
    for epoch in range(epochs):
        w = w - learning_rate*Fgradient_loss(w)
        w_loss = Floss(w)
        print(f"{w_loss} - {w}")
    
gradient_descent(loss, gradient_loss, init_weights())

```
